```python
# -*- coding: utf-8 -*-
"""
Created on Sat Nov  8 14:52:20 2014
 
@author: Pseudo
"""
import mechanicalsoup as ms
import re
import csv
import os
br = ms.Browser()
regex = re.compile(":|<.*?>|(\\n)|", re.IGNORECASE)
types = ["online", "paper"]
versions = ["standard", "modern_two", "modern_one", "legacy_two", "legacy_one"]
for type in types:
        for version in versions:
                initial_url = "http://www.mtggoldfish.com/prices/"+type+"/"+version+"/"
                if not os.path.exists("MTG_prices/"+type+"/"+version+"/"):
                        os.makedirs("MTG_prices/"+type+"/"+version+"/")
                page=br.get(initial_url)
                #collects list of urls to be navigated by sr
                html=page.soup
                frames=html.find_all("div", { "class" : "priceList-set" })
                for set in frames:
                        setName = set.find_all("a", { "class" : "priceList-set-header-link" })
 
                        for i in range(len(setName)):
                                setName[i]= regex.sub("", unicode(setName[i]))
                        setName = filter(None, setName)
                        if(len(setName) > 0):
                                print(setName)
                                rarity = set.find_all("h4", { "class" : "priceList-rarity-header" })
                                for i in range(len(rarity)):
                                        rarity[i]= regex.sub("", unicode(rarity[i]))
                                rarity = filter(None, rarity)
                                prices = set.find_all("dl", { "class" : "priceList-prices" })
                                filename = "MTG_prices/"+type+"/"+version+"/"+setName[0]+".txt"
                                with open(filename, 'w') as fp:
                                                for i in range(len(rarity)):
                                                        fp.write("{0:<35}{1:>10}{2:>10}\n\r".format(rarity[i], "Single", "Playset"))
                                                        priceListNames = prices[i].find_all("dt", { "class" : "priceList-prices-card" })
                                                        for ind in range(len(priceListNames)):
                                                                priceListNames[ind]= regex.sub("", unicode(priceListNames[ind]))
                                                        priceListNames = filter(None, priceListNames)
                                                        priceListPrices = prices[i].find_all("dd", { "class" : "priceList-prices-price" })
                                                        for ind in range(len(priceListNames)):
                                                                priceListPrices[ind]= float(regex.sub("", unicode(priceListPrices[ind])))
                                                        priceListPrices = filter(None, priceListPrices)
                                                        for index in range(len(priceListNames)):
                                                                fp.write("{0:<35}{1:>10}{2:>10}\n\r".format(priceListNames[index] , priceListPrices[index], 4* priceListPrices[index] ))
                                                        fp.write("\n\r")
```
